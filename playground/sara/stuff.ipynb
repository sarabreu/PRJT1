```python
import pandas as pd
import openpyxl
import tempfile


data_wb = openpyxl.load_workbook('/Users/saraabreu/Desktop/TRABALHOS/ISEP/PROJETO1/PRJT1_2022-2023_Sched_instance_1.xlsx')
shift_wb = data_wb.worksheets[0]

# Salva a planilha num arquivo temporário
with tempfile.NamedTemporaryFile(delete=False) as temp:
    data_wb.save(temp.name)
    # Lê os dados da planilha em um DataFrame
    shift_data = pd.read_excel(temp.name)

# Remove o arquivo temporário
os.unlink(temp.name)

shifts = {}

for i in range(shift_data.shape[0]):
    shift = shift_data.iloc[i]["Shift"]
    duration = shift_data.iloc[i]["Duration(min)"]
    setup_teams = shift_data.iloc[i]["Setup_teams"]
    log_train = shift_data.iloc[i]["Log_train(packs/hour)"]
    operators = shift_data.iloc[i]["Operators"]

    shifts[shift] = (duration, setup_teams, log_train, operators)

print(shifts)


# initial_data = wb.worksheets[1]
# instance_of_orders = wb.worksheets[2]
# setup_data = wb.worksheets[3]


```
